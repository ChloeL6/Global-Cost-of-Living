```python
# Dropping all rows with NaN values
cost_of_living.dropna(inplace=True)
cost_of_living
```

```python
# Function that will add all the column values together
def sum_columns(row):
  goods_sum = row.Meal + row.Meal_For_2 + row.McDonalds_Meal + row["Domestic_Beer_Rest(0.5L)"] + row['Imported_Beer_Rest(0.33L)'] + row.Cappuccino + row['Coke/Pepsi'] + row['Water_Rest(0.33L)'] + row.Milk + row.Rice + row.Eggs + row.Cheese + row.Chicken_Fillets + row.Beef + row.Apple + row.Bananas + row.Oranges + row.Tomato + row.Potato + row.Onion + row.Lettuce + row['Water_Market(1.5L)'] + row.Wine_Market + row['Domestic_Beer_Market(0.5L)'] + row['Imported_Beer_Market(0.33L)'] + row.Cigarettes
  return goods_sum

cost_of_living['goods_sum'] = cost_of_living.apply(sum_columns, axis=1)

cost_of_living['goods_sum']
```

```python
# Sorts goods sum column to show least expensive cities
cost_of_living.goods_sum.sort_values().head(10)
```

```python
# Sorts goods sum column to show most expensive cities
cost_of_living.goods_sum.sort_values(ascending=False).head(10)
```

```python
# Calculate the average of cost of goods
goods_len = cost_of_living.goods_sum.count()
goods_sum = cost_of_living.goods_sum.sum()

goods_average = round(goods_sum/goods_len, 2)
print(goods_average)
```
